# Scaling Normalization - PCA

In this section I will learn how to scale datas.

```python
import pandas as pd
```

```python
df = pd.read_csv("pima-indians-diabetes.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import normalize, scale
```

```python
x = df.drop("Outcome", axis=1)
```

```python
x.head()
```

```python
x2 = scale(x)
```

```python
pd.DataFrame(x2)
```

```python
norm = normalize(x)
```

```python
norm
```

```python
pd.DataFrame(norm)
```

PCA - Principle Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(4)
```

```python
x.head()
```

```python
x3 = pca.fit_transform(x)
```

```python
pd.DataFrame(x3,columns=["PCA1", "PCA2", "PCA3","PCA4"])
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```
