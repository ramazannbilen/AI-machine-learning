# "Human Learning" with iris data

Can you predict the species of an iris using petal and sepal measurements?

TASKS:
1. Read iris data into a pandas DataFrame, including column names.
2. Gather some basic information about the data.
3. Use groupby, sorting, and/or plotting to look for differences between species.
4. Come up with a set of rules that could be used to predict species based upon measurements.

BONUS: Define a function that accepts a row of data and returns a predicted species.
Then, use that function to make predictions for all existing rows of data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Task 1

```python
# read the iris data into a pandas DataFrame, including column names
df = pd.read_csv('iris.csv')
df
```

## Task 2

```python
# gather basic information
df.info()
df.describe()
```

## Task 3

```python
# use groupby to look for differences between the species 
df_gb = df.groupby('species')
print("\n mean is \n",df_gb.mean())
print("\n max is \n",df_gb.max())
print("\n min is \n",df_gb.min())
print("\n median is \n",df_gb.median())
print("\n std is \n",df_gb.std())
```

```python
# use sorting to look for differences between the species
df_sv = df.sort_values(by='species')
df_sv
```

```python
# Plotting mean values for each species
df_gb.mean().plot(kind='bar', figsize=(12,5), title='Mean Values by Species')
plt.ylabel('Mean')
plt.show()
```

```python
# map species to a numeric value so that plots can be colored by category
import seaborn as sns

mapping = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
df['num'] = df['species'].map(mapping)

sns.pairplot(df, hue='num', markers=["o", "s", "D"], palette='Set1')
plt.suptitle('Iris Veri Setinin Çift Grafikleri (Türe Göre Renklendirilmiş)', y=1.02)
plt.show();
```

## Bonus

```python
# define function that accepts a row of data and returns a predicted species

def predict_species(row):
    if row['petal_length'] < 2.0:
        return 'setosa'
    elif row['petal_width'] > 1.6:
        return 'virginica'
    else:
        return 'versicolor'
 
df['predicted_species'] = df.apply(predict_species, axis=1)

df.sample(20)

```

```python

setosa = df[df.species == 'setosa']
versicolor = df[df.species == 'versicolor']
virginica = df[df.species == 'virginica']
 
 
accuracy_setosa = (setosa['species'] == setosa['predicted_species']).mean()
accuracy_versicolor = (versicolor['species'] == versicolor['predicted_species']).mean()
accuracy_virginica = (virginica['species'] == virginica['predicted_species']).mean()
 
accuracy = (df['species'] == df['predicted_species']).mean()

print(f"General Prediction Accuracy: {accuracy * 100:.2f}%")

print(f"setosa Prediction Accuracy: {accuracy_setosa * 100:.2f}%")

print(f"versicolor Prediction Accuracy: {accuracy_versicolor * 100:.2f}%")

print(f"virginica Prediction Accuracy: {accuracy_virginica * 100:.2f}%")
```
